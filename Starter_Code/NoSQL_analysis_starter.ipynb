### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": {'$regex':"London"},
    "RatingValue": {"$gte": 4}
}

result = establishments.find(query)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a RatingValue greater than 4: {count}")
```

```python
# Display the first document in the results using pprint
result = establishments.find(query)
pprint(result[0])
```

```python
# Convert the result to a Pandas DataFrame
# Display the number of rows in the DataFrame
# Display the first 10 rows of the DataFrame
query = {
    "LocalAuthorityName": {'$regex':"London"},
    "RatingValue": {"$gte": 4}
}
results = establishments.find(query)

# Convert the cursor result to a list of dictionaries
documents = list(results)

# Create a Pandas DataFrame from the list of dictionaries
rating_vale_gte4_df = pd.DataFrame(documents)

# Display the number of rows in the DataFrame
num_rows = rating_vale_gte4_df.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(rating_vale_gte4_df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
penang_flavours_coords = {"latitude": 51.49014200,
"longitude": 0.08384000}

query = {'RatingValue':{'$regex': 5}}
# Capture the results to a variable
results = establishments.find(query)
# Define the aggregation pipeline
pipeline = [
    {
        "$match": {
            "RatingValue": 5,
            "geocode.latitude": {
                "$gte": penang_flavours_coords["latitude"] - degree_search,
                "$lte": penang_flavours_coords["latitude"] + degree_search,
            },
            "geocode.longitude": {
                "$gte": penang_flavours_coords["longitude"] - degree_search,
                "$lte": penang_flavours_coords["longitude"] + degree_search,
            },
        }
    },
    {
        "$addFields": {
            "Distance": {
                "$sqrt": {
                    "$sum": [
                        {"$pow": [{"$subtract": ["$geocode.longitude", penang_flavours_coords["longitude"]]}, 2]},
                        {"$pow": [{"$subtract": ["$geocode.latitude", penang_flavours_coords["latitude"]]}, 2]}
                    ]
                }
            }
        }
    },
    {
        "$sort": {"scores.Hygiene": 1, "Distance": 1}
    },
    {
        "$limit": 5
    }
]

# Execute the aggregation pipeline
result = list(establishments.aggregate(pipeline))

# Create a Pandas DataFrame from the result
top_5_df = pd.DataFrame(result)

# Display the top 5 establishments
print("Top 5 establishments with RatingValue 5, sorted by lowest hygiene score and nearest to 'Penang Flavours':")
print(top_5_df)


```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
# Print the number of documents in the result
# Print the first 10 results
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}
    }
]

# Execute the aggregation pipeline
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
num_documents = len(result)
print(f"Number of documents in the result: {num_documents}")

# Print the first 10 results
print("First 10 results:")
for i, doc in enumerate(result[:10], 1):
    print(f"{i}. Local Authority: {doc['_id']}, Count: {doc['count']}")
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
# Convert the result to a Pandas DataFrame
hygiene_score_0 = pd.DataFrame(result)

# Display the number of rows in the DataFrame
num_rows = hygiene_score_0.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(hygiene_score_0.head(10))
```
